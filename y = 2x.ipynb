```python
import pandas as pd
import matplotlib.pyplot as plt
```

Create a dataframe

```python
data = { 'x': [1, 2, 3, 4, 5],
        'y': [1, 4, 9, 16, 25]}
df = pd.DataFrame(data)
```

Create a scatter plot

```python
df.plot(kind='scatter', x='x', y='y')
```

Add a line fit to the scatter plot

```python
plt.plot(df['x'], df['y'], color='red', linestyle='--')
```

Display the plot


```python
plt.show()
```
